```python
new_names=[]
for name in names:
    if len(name.split())>2:
        new_names.append(name.split()[0]+'_'+name.split()[2])
    elif len(name.split())==2:
            new_names.append(name.split()[0]+'_'+name.split()[1])
#        print name.split(),len(name.split())
    else:
        new_names.append(name)
        
```
